```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import requests #web에서 가져올때

page = requests.get('https://www.basketball-reference.com/leagues/NBA_2019_per_game.html') #게임당 스텟지
total = page.text
#print(total)
```

```python
stats = total.split('<tbody>')[1].split('</tbody>')[0] #body 부분만 자르기
players = stats.split('<a href="/players/') #각 플레이어별 한줄

name = [] #선수이름
pos = [] #선수포지션
age = [] #선수나이
team = [] #소속팀

category  = ['"g" >','"gs" >','"mp_per_g" >','"fg_per_g" >', '"fga_per_g" >','"fg_pct" >','"fg3_per_g" >','"fg3a_per_g" >','"fg3_pct" >','"fg2_per_g" >','"fg2a_per_g" >','"fg2_pct" >','"efg_pct" >','"ft_per_g" >','"fta_per_g" >','"ft_pct" >','"orb_per_g" >','"drb_per_g" >','"trb_per_g" >','"ast_per_g" >','"stl_per_g" >',
                 '"blk_per_g" >','"tov_per_g" >','"pf_per_g" >','"pts_per_g" >']
obj = {}
for i in range(len(category)):
    obj[str(i)] = []

for i in range(1, len(players)):
    
    first = players[i].split('.html">',1)[1].split('<',1) #name 부분부터 끝까지
    name.append(first[0])
    
    second = first[1].split('"pos" >')[1].split('</td>',1) #pos 부분부터 끝까지
    pos.append(second[0])
    
    third = second[1].split('"age" >')[1].split('</td>',1) #age 부분부터 끝까지
    age.append(third[0])
    
    if 'html">' in third[1]:
        fourth = third[1].split('html">',1)[1].split('<',1) #team 부분부터 끝까지
        team.append(fourth[0])
    else:
        fourth = third[1].split('team_id" >',1)[1].split('<',1)
        team.append(fourth[0])   
    
    for i in range(len(category)):
        x = fourth[1].split(category[i],1)[1].split('</td>',1)
        obj[str(i)].append(x[0])
```

```python
data = {'Player': name, 'Pos': pos, 'Age': age, 'Tm': team, 'G': obj['0'], 'GS': obj['1'], 'MP': obj['2'], 
        'FG': obj['3'], 'FGA': obj['4'], 'FG%': obj['5'], '3P': obj['6'], '3PA': obj['7'], '3P%': obj['8'],
        '2P': obj['9'], '2PA': obj['10'], '2P%': obj['11'], 'eFG': obj['12'], 
        'FT' : obj['13'], 'FTA' : obj['14'], 'FT%' :obj['15'], 'ORB': obj['16'], 'DRB':obj['17'], 'TRB':obj['18'],
        'AST':obj['19'], 'STL':obj['20'],'BLK':obj['21'], 'TOV':obj['22'], 'PF':obj['23'], 'PTS':obj['24']
        }

df0 = DataFrame(data)
df0.to_excel("NBA stat.xlsx") 
df0
```
