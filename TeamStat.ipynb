```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import requests #web에서 가져올때

page = requests.get('https://www.basketball-reference.com/leagues/NBA_2019_ratings.html') #팀당 스텟지
total = page.text
#print(total)
```

```python
stats = total.split('<tbody>')[1].split('</tbody>')[0] #body 부분만 자르기
teams = stats.split('<a href="/teams/') #각 팀별 한줄

team = []
conf = []
div = []

category  = ['"wins" >','"losses" >','"win_loss_pct" >','"mov" >','"off_rtg" >','"def_rtg" >','"net_rtg" >']
obj = {}
for i in range(len(category)):
    obj[str(i)] = []

for i in range(1, len(teams)):
    first = teams[i].split('.html">',1)[1].split('<',1) #name 부분부터 끝까지
    team.append(first[0])
    
    second = first[1].split('"conf_id"')[1].split('" >',1)[1].split('</td>',1) #pos 부분부터 끝까지
    conf.append(second[0])
    
    third = second[1].split('"div_id"')[1].split('" >',1)[1].split('</td>',1) #age 부분부터 끝까지
    div.append(third[0])
    
    for i in range(len(category)):
        x = first[1].split(category[i],1)[1].split('</td>',1)
        obj[str(i)].append(x[0])
```

```python
data = {'Team': team, 'Conf':conf, 'Div':div, 'W': obj['0'], 'L': obj['1'], 
        'W/L%': obj['2'], 'MOV': obj['3'], 'ORtg': obj['4'], 'DRtg': obj['5'], 'NRtg': obj['6']
        }

df0 = DataFrame(data)
df0.to_excel("Team stat.xlsx") 
df0
```
