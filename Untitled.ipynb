```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as smapi
from sklearn.linear_model import LinearRegression
from pandas import Series, DataFrame
```

```python
df = pd.read_excel('Book2.xlsx') #각팀 성적
df.rename(columns={'WIN%':'WINP',
                   'FG%':'FGP',
                   '3PM':'TPM',
                   '3PA':'TPA',
                   '3P%':'TPP',
                   'FT%':'FTP',
                   '+/-':'PM'},inplace=True)
df.head()
```

```python
lm = smapi.ols(formula='WINP~PTS+FGM+FGA+FGP+TPM+TPA+TPP+FTM+FTA+FTP+OREB+DREB+REB+AST+TOV+STL+BLK+BLKA+PF+PFD+PM',data=df).fit() #OLS방법
lm.summary()
```

```python
import seaborn as sns
%matplotlib inline
sns.pairplot(df, x_vars=['PTS', 'FGM', 'FGA', 'FGP',
                         'TPM', 'TPA', 'TPP', 'FTM',
                         'FTA', 'FTP', 'OREB', 'DREB',
                         'REB', 'AST','TOV', 'STL', 
                         'BLK', 'BLKA', 'PF', 'PFD', 'PM'],y_vars='WINP', size=20,aspect= 0.7, kind='reg')
sns.lmplot('PTS','WINP',df)
```

```python
dfs = pd.read_excel('NBA stat.xlsx') #각팀 성적
dfs.head()
```

```python
player = []
team = []
stat = []
game = []
time = []
sstat = []

espn = []
espnnet = []
for i in range(len(dfs)):
    if not dfs['Tm'][i] == 'TOT':
        player.append(dfs['Player'][i])
        team.append(dfs['Tm'][i])
        game.append(dfs['G'][i])
        time.append(dfs['MP'][i])
        a = (dfs['PTS'][i]+dfs['TRB'][i]+dfs['AST'][i]+dfs['STL'][i]+dfs['BLK'][i]-dfs['TOV'][i])
        b = (dfs['PTS'][i]+dfs['TRB'][i]+1.4*dfs['AST'][i]+dfs['STL'][i]+1.4*dfs['BLK'][i]-0.7*dfs['TOV'][i]
            + dfs['FG'][i] + .5*(dfs['3P'][i]) -.8*(dfs['FGA'][i]-dfs['FG'][i]) + .25*dfs['FT'][i] - .8*(dfs['FTA'][i]-dfs['FT'][i]))
        stat.append(a)
        sstat.append(a*dfs['G'][i]/82*dfs['MP'][i]/48)
        espn.append(b)
        espnnet.append(b*dfs['G'][i]/82*dfs['MP'][i]/48)
            
#MPer = PTS + REB + AST + STL + BLK - TOV
#TMPer = MPer * G / 82 * MP / 48
    
#ESPN rating = PTS + REB + 1.4*AST + STL + 1.4*BLK -.7*TO + FGM + .5*TGM -.8*(FGA-FGM) + .25*FTM - .8*(FTA-FTM)
```

```python
data = {'Player': player, 'team': team, 'G':game, 'MP':time, 'MPer': stat, 'TMPer': sstat, 'ESPN':espn, 'TESPN': espnnet}

dfper = DataFrame(data)
dfper.to_excel("NBA MPer.xlsx")
dfper
```
